```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# Importing the dataset
train = pd.read_csv("train_ready_for_model_loan.csv")
train.head()
```

```python
train.shape
```

```python
X=train.drop(columns=['Loan_Status','Loan_ID']).values
y = train.iloc[:,6].values
print(X)
print(y)
```

```python
test = pd.read_csv("test_ready_for_model_loan.csv")
test.head()
```

```python
X_test=test.drop(columns=['Loan_ID']).values
print(X_test)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X)
X_test=sc.transform(X_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
classifier=Sequential()
```

```python
# Adding the input layer and the first hidden layer
classifier.add(Dense(output_dim=20,init='uniform',activation='relu',input_dim=20))
classifier.add(Dropout(0.6))
# Adding second hidden layer
classifier.add(Dense(output_dim=60,init='uniform',activation='relu'))
classifier.add(Dropout(0.6))
classifier.add(Dense(output_dim=30,init='uniform',activation='relu'))
classifier.add(Dropout(0.6))

classifier.add(Dense(output_dim=80,init='uniform',activation='relu'))
classifier.add(Dropout(0.6))

classifier.add(Dense(output_dim=60,init='uniform',activation='relu'))
classifier.add(Dropout(0.6))

#Adding the output layer
classifier.add(Dense(output_dim=1,init='uniform',activation='sigmoid'))
```

```python
classifier.add?
```

```python
#Compiling nueral network
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
print(X.shape)
print(X_test.shape)
print(y.shape)
#print(y_test.shape)
```

```python
classifier.fit(X,y,batch_size=10,nb_epoch=100)
```

```python
from sklearn.metrics import accuracy_score
y_pred=classifier.predict(X_train)
y_pred=(y_pred>0.5)
accuracy_score(y_pred,y_train)
```

```python
target = 'Loan_Status'
IDcol = ['Loan_ID']
#Predicting the test results
y_pred=classifier.predict(X_test)
y_pred=(y_pred>0.7)
test[target]=y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```
